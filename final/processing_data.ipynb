```python
import pandas as pd
import numpy as np
df =pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df_extract = df[["Order ID", "Category"]]
```

```python
gender_probs = {

    "Furniture": {"Female": 0.62, "Male": 0.38},

    "Technology": {"Male": 0.76, "Female": 0.24},

    "Office Supplies": {"Female": 0.73, "Male": 0.27}

}



# Assign gender based on probabilities

def assign_gender(category):

    probs = gender_probs[category]

    return np.random.choice(list(probs.keys()), p=list(probs.values()))



df["Gender"] = df["Category"].apply(assign_gender)
```

```python
# Young adult (18-25 age), Adult (26-44 age), Middle-age (45-59 age), Old age (60 age) 
age_group_probs = {

    "Furniture": {"Young adult": 0.12, "Adult": 0.24,"Middle-age":0.42, "Old age":0.22},

    "Technology": {"Young adult": 0.38, "Adult": 0.31,"Middle-age":0.18, "Old age":0.13},

    "Office Supplies": {"Young adult": 0.18, "Adult": 0.32,"Middle-age":0.28, "Old age":0.22}

}



# Assign gender based on probabilities

def assign_age_group(category):

    probs = age_group_probs[category]

    return np.random.choice(list(probs.keys()), p=list(probs.values()))



df["Age Group"] = df["Category"].apply(assign_age_group)
```

```python
# Young adult (18-25 age), Adult (26-44 age), Middle-age (45-59 age), Old age (60 age) 
cancel_probs = {

    "South": {"Return": 0.01, "Received": 0.99},

    "West": {"Return": 0.1, "Received": 0.9},

    "East": {"Return": 0.02, "Received": 0.98},
    
    "Central":{"Return": 0.03, "Received": 0.97}
}



# Assign gender based on probabilities

def assign_cancel_probs(category):

    probs = cancel_probs[category]

    return np.random.choice(list(probs.keys()), p=list(probs.values()))



df["Status"] = df["Region"].apply(assign_cancel_probs)
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Sort by 'Order Date'
df = df.sort_values('Order Date').reset_index(drop=True)

# Create a probability of "Online" increasing over time
df['Online Probability'] = np.linspace(0.1, 0.5, len(df))

# Assign "Online" or "Offline" based on the probability
df['Order Type'] = df['Online Probability'].apply(lambda x: 'Online' if np.random.rand() < x else 'Offline')

# Drop the 'Online Probability' column if not needed
df = df.drop(columns=['Online Probability'])
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Sort by 'Order Date'
df = df.sort_values('Order Date').reset_index(drop=True)

# Create a probability of "Online" increasing over time
df['Payment Probability'] = np.linspace(0.3, 0.7, len(df))

# Assign "Online" or "Offline" based on the probability
df['Payment Type'] = df['Payment Probability'].apply(lambda x: 'E-money' if np.random.rand() < x else 'Cash')

# Drop the 'Online Probability' column if not needed
df = df.drop(columns=['Payment Probability'])
```
